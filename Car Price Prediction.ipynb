```python
df_required = df.drop(["Car_Name"],axis = 1)
```

```python
df_required["no_of_years"] = 2021 - df["Year"]
```

```python
df_required.head()
```

```python
df_required.drop(["Year"], axis = 1, inplace = True)
```

```python
df_required = pd.get_dummies(df_required, drop_first = True)
```

```python
import seaborn as sns
sns.pairplot(df_required)
```

```python
#x = independent, y = dependent
independent = df_required.iloc[:,1:]
dependent = df_required.iloc[:,0]
```

```python
independent.head()
```

```python
dependent.head()
```

```python
# Feature Importance Heirarchy
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(independent, dependent)
```

```python
model.feature_importances_
```

```python
feature_importance = pd.Series(model.feature_importances_, index = independent.columns)
feature_importance.nlargest(5).plot(kind = 'barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
independent_train, independent_test, dependent_train, dependent_test = train_test_split(independent, dependent, test_size = 0.2)
```

```python
independent_train.head()
```

```python
independent_train.shape
```

```python
dependent_train.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_random = RandomForestRegressor()
```

```python
import numpy as np
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
n_estimators
```

```python
#Hyperparameter Tuning
from sklearn.model_selection import RandomizedSearchCV

n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
max_features = ["auto", "sqrt", "log2"]
max_depth = [int(x) for x in np.linspace(start = 5, stop = 30, num = 6)]
min_samples_split = [2, 5, 25, 50, 100]
min_samples_leaf = [1, 3, 7]
```

```python
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

```python
rf = RandomForestRegressor()
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 10, scoring = "neg_mean_squared_error", n_jobs = 1, cv = 5 , verbose = 5, random_state = 42)
```

```python
rf_random.fit(independent_train,dependent_train)
```

```python
rf_random.best_params_
```

```python
predictions=rf_random.predict(independent_test)
```

```python
sns.distplot(dependent_test-predictions)
```

```python
from sklearn.metrics import r2_score
r2_score(dependent_test, predictions)
```

```python
import pickle
# open a file, where you ant to store the data
file = open('random_forest_regression_model.pkl', 'wb')

# dump information to that file
pickle.dump(rf_random, file)
```
